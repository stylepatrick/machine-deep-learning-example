```python
with open("/content/drive/MyDrive/NeuralNetwork/verwandlung.txt", "r", encoding="utf-8") as file:
  text = file.read()

text = "\n".join(text.split("\n")[59:1952])

unique_chars = set(text)
int_to_char = {}
char_to_int = {}

for i, j in enumerate(unique_chars):
    int_to_char[i] = j
    char_to_int[j] = i
```

```python
print(int_to_char[0])
print(char_to_int[','])
```

```python
length = 40

x = []
y = []

for i in range(0, len(text) - length):
  line = text[i:i+length]
  line_in_int = []
  for j in range(0, len(line)):
    line_in_int.append(char_to_int[line[j]])
  x.append(line_in_int)
  char = text[i+length]
  y.append(char_to_int[char])

print(x[0])
print(y[0])
```

```python
# encode
txt = []
for p in range(0, len(x[0])):
  txt.append(int_to_char[x[0][p]])
txt
```

```python
from keras.utils import to_categorical

x_train = to_categorical(x, num_classes=len(unique_chars))
y_train = to_categorical(y, num_classes=len(unique_chars))
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM

model = Sequential()
model.add(LSTM(128, input_shape=(40, 68)))
model.add(Dense(68, activation="softmax"))

model.compile(optimizer="adam", loss="categorical_crossentropy")
```

```python
from keras.callbacks import ModelCheckpoint

save_model = ModelCheckpoint("weights.{epoch:02d}-{loss:.2f}.hdf5")
```

```python
model.fit(x_train, y_train, batch_size=32, epochs=10, callbacks=[save_model])
```

```python
from keras.models import load_model

model.save("/content/drive/MyDrive/NeuralNetwork/weights.hdf5")
model = load_model("/content/drive/MyDrive/NeuralNetwork/weights.hdf5")

model.summary()
```

```python
import numpy as np

predict_txt = text[200:240]
predict_int = []
for i in range(0, 40):
  predict_int.append(char_to_int[predict_txt[i]])

prediction = to_categorical(predict_int, num_classes=len(unique_chars))
prediction = prediction.reshape(1, 40, 68)
res = model.predict(prediction)
print(int_to_char[np.argmax(res[0])])
print(text[200:241])
```
